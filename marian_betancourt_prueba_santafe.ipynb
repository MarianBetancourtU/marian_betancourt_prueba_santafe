```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("C:/Users/Lenovo/Desktop/Prueba tecnica Santa Fe energy/GeneraciónCauya2024.csv")
```

```python
datos.head()
```

```python
datos_g = datos.drop(['Unnamed: 0'], axis=1)
datos_g = datos_g.fillna(0)
datos_g.head()
```

```python
datos_g.info()
```

```python
datos_g['Total_Generada'] = datos_g[['Values_Hour01', 'Values_Hour02', 'Values_Hour03', 'Values_Hour04', 
    'Values_Hour05', 'Values_Hour06', 'Values_Hour07', 'Values_Hour08',
    'Values_Hour09', 'Values_Hour10', 'Values_Hour11', 'Values_Hour12',
    'Values_Hour13', 'Values_Hour14', 'Values_Hour15', 'Values_Hour16',
    'Values_Hour17', 'Values_Hour18', 'Values_Hour19', 'Values_Hour20',
    'Values_Hour21', 'Values_Hour22', 'Values_Hour23', 'Values_Hour24']].sum(axis=1)
datos_g.head()
```

```python
# Convertir la columna de fechas al formato datetime automáticamente
datos_g['Date'] = pd.to_datetime(datos_g['Date'])

# Agrupar por mes y calcular el total
datos_g['Mes'] = datos_g['Date'].dt.to_period('M')  # Extraer el mes (como Periodo)
total_mes = datos_g.groupby('Mes')['Total_Generada'].sum().reset_index()

# Mostrar el resultado
total_mes

# Usé la ayuda de chat gpt para esta parte
```

```python
ultimos_30_dias = datos_g.sort_values(by='Date', ascending=False)
ultimos_30_dias = ultimos_30_dias.head(30)
ultimos_30_dias
```

```python
estadísticas_datos = datos_g.describe()
estadísticas_datos
```

```python
estadísticas_datos = estadísticas_datos.head(24)
estadísticas_datos = estadísticas_datos.transpose()
estadísticas_datos
```

```python
estadísticas_datos["Values_Hour"] =range(1, len(estadísticas_datos) + 1)
estadísticas_datos
```

```python
# 1. Energia generada ultimos 30 dias 
sns.barplot(data= ultimos_30_dias, x="Date" , y ="Total_Generada")
plt.xticks(rotation=90)
plt.show()
```

```python
# 2. Energia generada por mes durante 2024
# 4. Distribución y que tan dispersa es cada mes: se evidencia que la producción de energia es mucho mayor en los meses a mitad del año que iniciando
# y terminando el año.
sns.barplot(data= total_mes, x="Mes" , y ="Total_Generada")
plt.xticks(rotation=90)
plt.show()
```

```python
# 3. Promedio de energia generada cada hora.
sns.barplot(data= estadísticas_datos, x='Values_Hour', y="mean")
plt.xticks(rotation=90)
plt.show()
```
